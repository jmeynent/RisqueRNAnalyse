```python
# Obtenez le chemin du répertoire actuel (qui est RisqueRNAnalyse)
current_dir = os.getcwd()

# Construisez le chemin complet vers votre fichier Excel
fichier_excel = os.path.join(current_dir, 'RisqueRNAnalyse', 'Database', 'Legislatives2022CircoT2.xlsx')

# Chargez le fichier Excel dans un DataFrame
Legi2022CircoT2 = pd.read_excel(fichier_excel)
```

```python
Legi2022CircoT2.columns
circo222 = Legi2022CircoT2[['Code du département', 'Code de la circonscription', 'Inscrits', 'Abstentions', '% Abs/Ins', 'Votants',
       '% Vot/Ins', 'Blancs', '% Blancs/Ins', '% Blancs/Vot', 'Nuls',
       '% Nuls/Ins', '% Nuls/Vot', 'Exprimés', '% Exp/Ins', '% Exp/Vot', 'Nuance', 'Voix', '% Voix/Ins',
       '% Voix/Exp', 'Sièges', 'Nuance 2', 'Voix 2', '% Voix/Ins 2', '% Voix/Exp 2', 'Sièges 2']]
circo222
```

```python
# Spécifiez les colonnes à exclure
colonnes_a_exclure = ['Code du département', 'Code de la circonscription', 'Code circonscription législative', 'Code département']

# Renommer les colonnes en ajoutant "22" sauf celles à exclure
circo222 = circo222.rename(columns={col: f"{col}22" for col in circo222.columns if col not in colonnes_a_exclure})
circo242 = circo242.rename(columns={col: f"{col}24" for col in circo242.columns if col not in colonnes_a_exclure})
```

```python
circo222 =  circo222.rename(columns={'Code du département' : 'Code département', 'Code de la circonscription' : 'Code de la circonscription législative'})
```

```python
circo222['Code de la circonscription législative'] = circo222['Code de la circonscription législative'].apply(lambda x: f"{x:02d}")

# Créer une nouvelle colonne en fusionnant 'Code du département' et 'Code de la circonscription législative'
circo222['Code circonscription législative'] = circo222['Code département'] + circo222['Code de la circonscription législative']
circo222
```

```python
circo222 = circo222.drop(['Code département', 'Code de la circonscription législative'], axis=1)
circo222
```

```python
merged_df = pd.merge(circo222, circo242, on='Code circonscription législative', how='outer')
merged_df
```

```python
# Créer la colonne 'bord22'
merged_df['bord22'] = np.where(merged_df['Sièges 222'] == 'Elu', merged_df['Nuance 222'], merged_df['Nuance22'])

# Créer la colonne 'bord24' en utilisant une fonction personnalisée
def determine_bord24(row):
    if row['Elu 124'] == 'élu':
        return row['Nuance candidat 124']
    elif row['Elu 224'] == 'élu':
        return row['Nuance candidat 224']
    elif row['Elu 324'] == 'élu':
        return row['Nuance candidat 324']
    elif row['Elu 424'] == 'élu':
        return row['Nuance candidat 424']
    else:
        return np.nan  # ou toute autre valeur par défaut

merged_df['bord24'] = merged_df.apply(determine_bord24, axis=1)

merged_df
```

```python
mapping_bord22 = {
    'ENS': 'ENS',
    'NUP': 'NUP',
    'RN': 'XD',
    'LR': 'LR',
    'DVG': 'DVG',
    'DVD': 'LR',
    'REG': 'REG',
    'DVC': 'DVC',
    'UDI': 'LR',
    'DSV': 'XD',
    'DIV': 'DIV'
}

# Remplacer les labels dans la colonne 'bord22' en utilisant le dictionnaire de correspondance
merged_df['bord22'] = merged_df['bord22'].replace(mapping_bord22)

merged_df['bord22'].value_counts()
```

```python
# Dictionnaire de correspondance des labels pour bord24
mapping_bord24 = {
    'ENS': 'ENS',
    'UG': 'UG',
    'RN': 'XD',
    'LR': 'LR',
    'DVD': 'LR',   # Regroupe DVD sous LR
    'UXD': 'XD',   # Regroupe UXD sous RN
    'DVG': 'DVG',
    'REG': 'REG',
    'HOR': 'ENS',  # Regroupe HOR sous ENS
    'DVC': 'DVC',
    'UDI': 'LR',   # Regroupe UDI sous LR
    'SOC': 'DVG',  # Regroupe SOC sous DVG
    'DIV': 'DIV',
    'ECO': 'DVG'   # Regroupe ECO sous DVG
}

# Fonction pour appliquer le remplacement
def apply_mapping(mapping, column):
    return column.replace(mapping, regex=True)

# Appliquer le remplacement à la colonne 'bord24'
merged_df['bord24'] = apply_mapping(mapping_bord24, merged_df['bord24'])


# Remplacer les labels dans la colonne 'bord24' en utilisant le dictionnaire de correspondance
merged_df['bord24'] = merged_df['bord24'].replace(mapping_bord24)
```

```python
merged_df['bord24'].value_counts()
```

```python
# Filtrer les données pour bord24='ENS'
filtered_data = merged_df[merged_df['bord24'] == 'ENS']

# Compter les occurrences de chaque valeur de bord22
counts_bord22 = filtered_data['bord22'].value_counts()

# Créer un histogramme
plt.figure(figsize=(10, 6))
plt.bar(counts_bord22.index, counts_bord22.values, color='yellow')
plt.xlabel('Bord du siège en 2022')
plt.ylabel('Nombre')
plt.title('Origine des sièges de la majorité présidentielle suite aux législatives 2024')
plt.show()
```

```python
# Filtrer les données pour bord24='ENS'
filtered_data = merged_df[merged_df['bord24'] == 'UG']

# Compter les occurrences de chaque valeur de bord22
counts_bord22 = filtered_data['bord22'].value_counts()

# Créer un histogramme
plt.figure(figsize=(10, 6))
plt.bar(counts_bord22.index, counts_bord22.values, color='red')
plt.xlabel('Bord du siège en 2022')
plt.ylabel('Nombre')
plt.title('Origine des sièges du nouveau front populaire suite aux législatives 2024')
plt.show()
```
