```python
results22 = pd.merge(clean_22T1, circo222, on='Code circonscription législative22', how='left') 
results22
```

```python
merged_df = pd.merge(results22, results24, left_on = 'Code circonscription législative22', right_on='Code circonscription législative', how='outer')
merged_df
```

```python
# Créer la colonne 'bord22'
merged_df['bord22'] = np.where(merged_df['Sièges 222'] == 'Elu', merged_df['Nuance 222'], merged_df['Nuance22'])

# Créer la colonne 'bord24' en utilisant une fonction personnalisée
def determine_bord24(row):
    if row['Elu 124'] == 'élu':
        return row['Nuance candidat 124']
    elif row['Elu 224'] == 'élu':
        return row['Nuance candidat 224']
    elif row['Elu 324'] == 'élu':
        return row['Nuance candidat 324']
    elif row['Elu 424'] == 'élu':
        return row['Nuance candidat 424']
    else:
        return np.nan  # ou toute autre valeur par défaut

merged_df['bord24'] = merged_df.apply(determine_bord24, axis=1)

merged_df
```

```python
mapping_bord22 = {
    'ENS': 'ENS',
    'NUP': 'NUP',
    'RN': 'XD',
    'LR': 'LR',
    'DVG': 'DVG',
    'DVD': 'LR',
    'REG': 'REG',
    'DVC': 'DVC',
    'UDI': 'LR',
    'DSV': 'XD',
    'DIV': 'DIV'
}

# Remplacer les labels dans la colonne 'bord22' en utilisant le dictionnaire de correspondance
merged_df['bord22'] = merged_df['bord22'].replace(mapping_bord22)

merged_df['bord22'].value_counts()
```

```python
# Dictionnaire de correspondance des labels pour bord24
mapping_bord24 = {
    'ENS': 'ENS',
    'UG': 'UG',
    'RN': 'XD',
    'LR': 'LR',
    'DVD': 'LR',   # Regroupe DVD sous LR
    'UXD': 'XD',   # Regroupe UXD sous RN
    'DVG': 'DVG',
    'REG': 'REG',
    'HOR': 'ENS',  # Regroupe HOR sous ENS
    'DVC': 'DVC',
    'UDI': 'LR',   # Regroupe UDI sous LR
    'SOC': 'DVG',  # Regroupe SOC sous DVG
    'DIV': 'DIV',
    'ECO': 'DVG'   # Regroupe ECO sous DVG
}

# Fonction pour appliquer le remplacement
def apply_mapping(mapping, column):
    return column.replace(mapping, regex=True)

# Appliquer le remplacement à la colonne 'bord24'
merged_df['bord24'] = apply_mapping(mapping_bord24, merged_df['bord24'])


# Remplacer les labels dans la colonne 'bord24' en utilisant le dictionnaire de correspondance
merged_df['bord24'] = merged_df['bord24'].replace(mapping_bord24)
```

```python
merged_df['bord24'].value_counts()
```

```python
# Filtrer les données pour bord24='ENS'
filtered_data = merged_df[merged_df['bord24'] == 'ENS']

# Compter les occurrences de chaque valeur de bord22
counts_bord22 = filtered_data['bord22'].value_counts()

# Créer un histogramme
plt.figure(figsize=(10, 6))
plt.bar(counts_bord22.index, counts_bord22.values, color='yellow')
plt.xlabel('Bord du siège en 2022')
plt.ylabel('Nombre')
plt.title('Origine des sièges de la majorité présidentielle suite aux législatives 2024')
plt.show()
```
